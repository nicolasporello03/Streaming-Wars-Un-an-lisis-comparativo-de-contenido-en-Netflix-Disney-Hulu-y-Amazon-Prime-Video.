```python
import pandas as pd
disney = pd.read_csv(r'C:\Users\nicol\OneDrive\Documentos\GitHub\Streaming_Wars\datos\disney_plus_titles.csv')
```

```python
disney.head()
```

```python
print(disney.isna().sum())
```

```python
#show_id
disney['show_id'] = disney['show_id'].apply(lambda x: x.replace('s', '').replace(' ', '') if 's' in x else x)
disney['show_id'] = disney['show_id'].astype('int')
disney.info()
```

```python
#type
disney.groupby('type').size()
```

```python
#director y tabla de Director
disney['director'] = disney['director'].fillna("Desconosidos")
Director_Disney = disney[['show_id','director']]
disney = disney.drop(columns=['director'])

Director_Disney['director'] = Director_Disney['director'].str.split(', ')
Director_Disney = Director_Disney.explode('director')

```

```python
#cast y creacion de tabla de la misma
disney['cast'] = disney['cast'].fillna("Desconosidos")
Cast_Disney = disney[['show_id','cast']]
disney = disney.drop(columns=['cast'])

Cast_Disney['cast'] = Cast_Disney['cast'].str.split(', ')
Cast_Disney = Cast_Disney.explode('cast')
Cast_Disney.groupby('cast').size().sort_values(ascending=False)
```

```python
#country
disney['country'] = disney['country'].fillna("Desconosidos")
country_Disney = disney[['show_id','country']]
disney = disney.drop(columns=['country'])

country_Disney['country'] = country_Disney['country'].str.split(', ')
country_Disney = country_Disney.explode('country')
country_Disney.groupby('country').size().sort_values(ascending=False).head()
```

```python
#date_added 
disney['date_added'] = disney['date_added'].ffill()
disney['date_added'].head()
disney['date_added'] = pd.to_datetime(disney['date_added'], format='mixed')

```

```python
#release_year
disney.groupby('release_year').size()
```

```python
#rating
disney['rating'] = disney['rating'].fillna('Desconosidos')
disney.groupby('rating').size()
```

```python
#duration
pelis_disney=disney[disney['duration'].str.contains('min', case=False, na=False)]
pelis_disney = pelis_disney[['show_id','duration']]
pelis_disney

```

```python
#series tabla
series_disney=disney[disney['duration'].str.contains('Season|Seasons', case=False, na=False)]
series_disney = series_disney[['show_id','duration']]
series_disney
disney = disney.drop(columns=['duration'])
series_disney['duration'] = series_disney['duration'].str.replace('Season|Seasons|s', '', regex=True)
series_disney['duration']= series_disney['duration'].astype('int')
```

```python
#listed_in tabla
listed_in_disney = disney[['show_id','listed_in']]
listed_in_disney['listed_in'] = listed_in_disney['listed_in'].str.split(', ')
listed_in_disney = listed_in_disney.explode('listed_in')
disney = disney.drop(columns=['listed_in'])
listed_in_disney.groupby('listed_in').size().sort_values(ascending=False)
```

```python
#descripcion 
disney = disney.drop(columns=['description'])
disney
```

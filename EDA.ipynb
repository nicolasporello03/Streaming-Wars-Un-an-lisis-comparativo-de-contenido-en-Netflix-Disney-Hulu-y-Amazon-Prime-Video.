AMAZON


```python
#combertimos los datos en entero y string
amazon['show_id'] = amazon['show_id'].str.replace('s', '', regex=False)
amazon['show_id'] = amazon['show_id'].astype('int')

#dividir las series con los minutos de la columna duracion
amazon['duration_min'] = amazon['duration'].apply(lambda x : int(x.split()[0]) if 'min' in x else 0)
amazon['duration_seasons'] = amazon['duration'].apply(lambda x : int(x.split()[0]) if 'Seasons' in x else 0)



#cambiamos valores que significan lo mismo solo que escritos de otra manera
amazon['rating'] = amazon['rating'].replace({'AGES_16_':'16+' , 'AGES_18_':'18+' , 'NOT_RATE':'UNRATE' , '16':'16+'})

#pasamos a tiempo la columna date_added
amazon['date_added'] = amazon['date_added'].ffill()
amazon['date_added'] = pd.to_datetime(amazon['date_added'], format='%B %d, %Y')

amazon['cast']=amazon['cast'].fillna('Desconocido')
amazon['country']=amazon['country'].fillna('Desconocido')
amazon['rating']=amazon['rating'].fillna('Desconocido')
amazon['director']=amazon['director'].fillna('Desconocido')
print(amazon.isna().sum())
```

NETFLIX

```python

nexflix['show_id'] = nexflix['show_id'].str.replace('s', '', regex=False)
nexflix['show_id'] = nexflix['show_id'].astype('int')
nexflix['director'] = nexflix['director'].fillna('Desconocido')
nexflix['cast'] = nexflix['cast'].fillna('Desconocido' )


nexflix['duration_min'] = nexflix['duration'].apply(lambda x: int(str(x).split()[0]) if isinstance(x, str) and 'min' in x else 0)
nexflix['duration_seasons'] = nexflix['duration'].apply(lambda x: int(str(x).split()[0]) if isinstance(x, str) and 'Seasons' in x else 0)


nexflix['country'] = nexflix['country'].fillna('Desconocido' )
nexflix['date_added'] = nexflix['date_added'].str.strip()
nexflix['date_added'] = pd.to_datetime(nexflix['date_added'], format='%B %d, %Y')
nexflix= nexflix.dropna()
print(nexflix.isna().sum())
nexflix.describe()
```

HULU


```python

hulu['show_id'] = hulu['show_id'].str.replace('s', '', regex=False)
hulu['show_id'] = hulu['show_id'].astype('int')

hulu.drop(['director','cast'], axis=1, inplace=True)

hulu['country'] = hulu['country'].fillna('Desconocido')

hulu['rating'] = hulu['rating'].fillna('Desconocido')


hulu['date_added'] = hulu['date_added'].ffill()
hulu['date_added'] = hulu['date_added'].str.strip()
hulu['date_added'] = pd.to_datetime(hulu['date_added'], format='%B %d, %Y')

#desonocidos los que tienen minutos o season  en rating
hulu['rating'] = hulu['rating'].apply(lambda x: 'Desconocido' if any(word in x for word in ['min', 'Season']) else x)



hulu['description'] = hulu['description'].fillna('Desconocido')
hulu['duration'] = hulu['duration'].str.replace(r'(min|Season|s)', '', regex=True)
hulu['duration'] = hulu['duration'].astype('float')
media_dur_hulu = hulu['duration'].mean()
hulu['duration'] = hulu['duration'].fillna(media_dur_hulu)
hulu.info()
```
